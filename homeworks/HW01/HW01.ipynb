```python
#1
import numpy as np
```

```python
#2
mass = np.array([1, 2, 3])
mass_mass = np.array([[11, 12, 13], [21, 22, 23], [31, 32, 33]])

zero_mass = np.zeros(5)
ones_mass = np.ones(5)

two_mass = np.arange(2, 10, 2)

print("Обычный масив:", mass)
print("Матрица 3x3:\n", mass_mass)
print("Массив нулей:", zero_mass)
print("Массив едениц:", ones_mass)
print("Масив четных(2 - 10):", two_mass)
```

```python
#3
print("Работа с mass_mass:\n")
print("Тип данных в масиве:", mass_mass.dtype)
print("Размерность масива:", mass_mass.shape)
print("Вложенность масива:", mass_mass.ndim)
```

```python
#4
print("Работа с mass_mass:\n")
print("Вывод всего масива:\n", mass_mass)
print("Вывод 2-3 строк:\n", mass_mass[1:3])
print("Вывод 1-2 столбцов из 1 строки:\n", mass_mass[0,0:2])
print("Вывод 3 столбца:\n", mass_mass[0:3,2:3])
```

```python
#5
print("Работа с mass:\n")
print("Среднее значение в масиве:", mass.mean())
print("Максимальное значение:", mass.max())
print("Индекс максимального значения:",mass.argmax())
print("Минимальное значение:", mass.min())
print("Индекс синимального значения:",mass.argmin())
```
